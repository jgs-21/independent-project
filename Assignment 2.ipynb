<h1><u>Site Location Problem</u></h1>

<h3><center>Locating a rock aggregate factory site depending upon geology, transport, and population.</center></h3>

<p>Here we will explore where the best areas are in the UK to site a rock aggregate factory. Three factors need to be considered, which are;</p>
    <ul>
        <li>good geology,</li>
        <li>the importance of transport, and</li>
        <li>the population workforce.</li>
    </ul>

<p>Below are three maps displaying the geology, transport, and population workforce of the UK. Note the lighter the area (the higher the value) the better the area for a site.</p>

<h3>Geology:</h3>
<p>Below is a map that displays areas of good geology in the UK - the lighter the area (higher the value) the better the geology</p>

```python
{
    "tags": [
        "hide-input",
    ]
}

# Import modules
import matplotlib
import matplotlib.pyplot
import matplotlib.widgets
import csv

# Load data
g = open('geology.txt', newline='')
g_reader = csv.reader(g, quoting=csv.QUOTE_NONNUMERIC)


# GEOLOGY:

# Create the empty lists (containers for the data) before any data is processed
geology = []  # list (container) for the environment

# Read each row in the data being loaded
for row in g_reader:
    # Create an empty list for the data to be added to
    g_rowlist = []
    # Before each row is processed, for each value in the row, add the value
    # to the empty list (rowlist)
    for values in row:
        # Add the values to the empty list (rowlist).
        g_rowlist.append(values)
    # After the row is processed, add the data to the environment
    geology.append(g_rowlist)
      
# Close the file after all the data is processed
g.close()

fig = matplotlib.pyplot.figure(figsize=(8,8))
ax = fig.add_axes([0, 0, 1, 1])

matplotlib.pyplot.ylim(530, 0)
matplotlib.pyplot.xlim(0, 330)

# The following code to remove the values from the plot axes was taken from
# https://stackoverflow.com/questions/19859609/how-to-remove-values-on-x-y-axis-on-plot-in-matplotlib/19859780
frame = matplotlib.pyplot.gca()
frame.axes.get_xaxis().set_ticks([])
frame.axes.get_yaxis().set_ticks([])

matplotlib.pyplot.imshow(geology, cmap='binary_r')
matplotlib.pyplot.title('Geology', size=16)
matplotlib.pyplot.colorbar()
matplotlib.pyplot.show()
```

```python
# Import modules
import matplotlib
import matplotlib.pyplot
import matplotlib.widgets
import csv
import ipywidgets as widgets
import numpy


# Load geology data
g = open('geology.txt', newline='')
g_reader = csv.reader(g, quoting=csv.QUOTE_NONNUMERIC)

# Load transport data
t = open('transport.txt', newline='')
t_reader = csv.reader(t, quoting=csv.QUOTE_NONNUMERIC)

# Load population data
p = open('population.txt', newline='')
p_reader = csv.reader(p, quoting=csv.QUOTE_NONNUMERIC)


# GEOLOGY:

# Create the empty lists (containers for the data) before any data is processed
geology = []  # list (container) for the environment

# Read each row in the data being loaded
for row in g_reader:
    # Create an empty list for the data to be added to
    g_rowlist = []
    # Before each row is processed, for each value in the row, add the value
    # to the empty list (rowlist)
    for values in row:
        # Add the values to the empty list (rowlist).
        g_rowlist.append(values)
    # After the row is processed, add the data to the environment
    geology.append(g_rowlist)
      
# Close the file after all the data is processed
g.close()


# TRANSPORT:

# Create the empty lists (containers for the data) before any data is processed
transport = []  # list (container) for the environment

# Read each row in the data being loaded
for row in t_reader:
    # Create an empty list for the data to be added to
    t_rowlist = []
    # Before each row is processed, for each value in the row, add the value
    # to the empty list (rowlist)
    for values in row:
        # Add the values to the empty list (rowlist).
        t_rowlist.append(values)
    # After the row is processed, add the data to the environment
    transport.append(t_rowlist)
      
# Close the file after all the data is processed
t.close()


# POPULATION:

# Create the empty lists (containers for the data) before any data is processed
population = []  # list (container) for the environment

# Read each row in the data being loaded
for row in p_reader:
    # Create an empty list for the data to be added to
    p_rowlist = []
    # Before each row is processed, for each value in the row, add the value
    # to the empty list (rowlist)
    for values in row:
        # Add the values to the empty list (rowlist).
        p_rowlist.append(values)
    # After the row is processed, add the data to the environment
    population.append(p_rowlist)
      
# Close the file after all the data is processed
p.close()

wg = 0
slider1 = widgets.FloatSlider(
    value=0,
    min=0,
    max=1,
    step=0.1,
    description='Geology Weight:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)
slider1.value = wg
def on_change(v):
    wg = v['new'] 
slider1.observe(on_change, names='value')
#display(slider1)



wp = 0
slider2 = widgets.FloatSlider(
    value=0,
    min=0,
    max=1,
    step=0.1,
    description='Population Weight:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)
slider2.value = wp
def on_change(v):
    wp = v['new'] 
slider2.observe(on_change, names='value')
#display(slider2)



wt = 0
slider3 = widgets.FloatSlider(
    value=0,
    min=0,
    max=1,
    step=0.1,
    description='Transport Weight:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)
slider3.value = wt
def on_change(v):
    wt = v['new'] 
slider3.observe(on_change, names='value')
#display(slider3)



result = []

for i in range(len(geology)):
    resultrow = []
    for j in range(len(geology[i])):
        value = geology[i][j]*wg + population[i][j]*wp + transport[i][j]*wt
        resultrow.append(value)
    result.append(resultrow)
    max_val = numpy.max(result)
    #print('Maximum value:', max_val)
    min_val = numpy.min(result)
    #print('Minimum value:', min_val)
    rescale_result = ((result- min_val) * (1/(max_val - min_val) * 255))
    


fig = matplotlib.pyplot.figure(figsize=(8,8))
ax = fig.add_axes([0, 0, 1, 1])

# Plot data in a graph
matplotlib.pyplot.ylim(530, 0)
matplotlib.pyplot.xlim(0, 330)

# The following code to remove the values from the plot axes was taken from
# https://stackoverflow.com/questions/19859609/how-to-remove-values-on-x-y-axis-on-plot-in-matplotlib/19859780
frame = matplotlib.pyplot.gca()
frame.axes.get_xaxis().set_ticks([])
frame.axes.get_yaxis().set_ticks([])


matplotlib.pyplot.imshow(rescale_result, cmap='binary_r')
matplotlib.pyplot.title('MCE', size=16)
matplotlib.pyplot.colorbar()
matplotlib.pyplot.show()

display(slider1)
display(slider2)
display(slider3)
```
